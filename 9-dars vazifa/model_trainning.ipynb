#ENCODING

```python
df_encoded=df.copy()
```

```python
for col in df_encoded.select_dtypes(include=['object']).columns:
    cardinality=df_encoded[col].nunique()
    if cardinality<=5:
        le=LabelEncoder()
        df_encoded[col]=le.fit_transform(df_encoded[col])
    else:
        df_encoded=pd.get_dummies(df_encoded,columns=[col],dtype=int,drop_first=True)
df=df_encoded
```

```python
df.info()
```

#SCALING

```python
num_col=df.select_dtypes(include=['int64','float64']).columns #bu kod datasetdagi hamma  ustunlardagi int va float ma'lumot turi bolgan ustunlarni ajratib oladi
```
